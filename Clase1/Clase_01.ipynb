## Bases de datos:
Antes de utilizar el programa R es recomendable revisar la base de datos. Para ello se puede abrir la base de datos mediante algún programa como excel o libre office y deben cumplirse que:
1. El nombre de las variables no tenga espacios,
    por ejemplo, una variable llamada *clase social* debe renombrarse como (por ejemplo) *clase_social*.
2. Comprobar que la base de datos esté completa, es decir, no tenga datos faltantes, observaciones imcompletas o errores de tipeo.

* **Importante!!**: En general, la base de datos será guardada en formato *csv* de esa forma facilita al programa leer las bases de datos.

El formato *.csv* es un formato de texto denominado 'coma separated values', lo que hace es crear un archivo de texto que divide las filas y columnas del excel a traves de puntos y comas. En este formato R es capaz de leer la base de datos de una forma rápida y sin errores.

Para guardar un archivo excel en formato *.csv* cuando colocas *guardar como* en la parte inferior del nombre debes buscar formato *.csv*

### Filtrar bases de datos
Algunas veces, nos serán de utilidad solo los datos que cumplan una condición particular. Para ello en R se utiliza la función `subset(data,condición)`, esta  entrega una base de datos filtrada con la condición impuesta. Por ejemplo en nuestra base de datos buscaremos los Pokemon con más de 160 puntos de `Attack`, para esto hacemos lo siguiente:

```r
fuertes  = subset(data, Attack>160)
fuertes
```

#### En R también existen las funciones clasicas de sumar, restar, multiplicar y dividir dadas por `+` `-` `*` `/`. En particular existen dos funciones `max` y `min` que entregan el máximo y el mínimo de un conjunto de datos. Por ejemplo, veamos cuales son los valores maximos de `Attack`, y minimos de `Defense` de la base de datos anterior


```r
max(data$Attack)
```

```r
min(data$Defense)
```

####  Con estos comandos, podemos buscar a los pokemon que tienen estos valores de `Attack` y `Defense`, respectivamente

```r
subset(data, Attack== max(data$Attack))
```

```r
subset(data, Defense== min(data$Defense))
```

#### Tambien es posible verificar condiciones logicas, como por ejemplo si un Pokemon es Legendario o no

```r
Legendarios = subset(data, Legendary== 'True')
head(Legendarios)
```

También podemos utilizar mas de una condición utilizando los conectores logicos **y** y **o** representados en R por `&` y `|` respectivamente. Por ejemplo podriamos buscar los pokemones con `HP` mayor que 130 **y** que `Type.1` sea `Water`.

```r
subset(data,HP>130 & Type.1 == 'Water')
```
